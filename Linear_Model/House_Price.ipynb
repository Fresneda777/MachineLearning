```python
import warnings
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
X = [[300], [500], [600], [800], [1200],[1500],[1750],[2000],[2200]]
y = [[80], [130], [180], [260], [390],[380],[450],[380],[420]]
```

```python
plt.figure()
plt.title('Precio de la casa según tamaño metros. ')
plt.xlabel('Tamaño de la casa.')
plt.ylabel('Precio de la casa en dólares')
plt.plot(X, y, 'k.')
plt.axis([0, 2500, 0, 500])
plt.grid(True)
plt.show()
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
plt.figure()
plt.title('Precio de la casa según tamaño metros. ')
plt.xlabel('Tamaño de la casa.')
plt.ylabel('Precio de la casa en dólares')
plt.plot(X, y, 'k.')
plt.plot(X, model.predict(X),'g')
plt.axis([0, 2500, 0, 500])
plt.grid(True)
plt.show()
```

```python
from ipywidgets import interact, Button
```

```python
def precioPronostico(size):
    print ('Una casa de tamaño %d metros cuesta cerca de: $%i000 dòlares'%(size,model.predict([[size]])))
```

```python
i = interact(precioPronostico, size=(0,3000))
```

```python
import numpy as np
print ('Residual suma de cuadrados: %.2f' % np.mean((model.predict(X)- y) ** 2))
```
